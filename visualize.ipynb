```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')
data = pd.read_csv('/content/twitter_training.csv')
print("DataFrame Columns:", data.columns)
print(data.head())
sia = SentimentIntensityAnalyzer()
def get_sentiment(text):
    sentiment_scores = sia.polarity_scores(text)
    return sentiment_scores['compound']
data['sentiment'] = data['Borderlands'].apply(get_sentiment)
print(data[['Borderlands', 'sentiment']].head())
def sentiment_category(score):
    if score >= 0.05:
        return 'Positive'
    elif score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'
data['sentiment_category'] = data['sentiment'].apply(sentiment_category)

```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='sentiment_category', palette='coolwarm')
plt.title('Sentiment Distribution in Social Media Data')
plt.xlabel('Sentiment Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
